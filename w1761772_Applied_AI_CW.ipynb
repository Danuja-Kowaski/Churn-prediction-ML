*Part A - Problem Domain*

__1.1 Churn Prediction in Telcom industries__

Telecom industries have become a highly competitive market due to the wide range of telecom services being provided to customers. Customers have multiple options with varying prices and other beneficial aspects, each depending on the service provider. Due to these reasons, companies can lose many valuable customers and this progression is known as Customer churn, one of the most common and severe threats to a company’s revenue. Customer churn in telecom industries can occur due to many reasons such as cost, poor network quality, and lack of customer support. Due to these reasons, telecom companies have now made reducing customer churn one of their main objectives by predicting high-risk customers and estimating their time of churn. Predicting customer churn involves analyzing customer data such as purchase history, usage patterns, and other relevant information to identify patterns or trends that will likely affect churn. Since huge amounts of customer data are being generated by these companies, automated analysis techniques can be utilized for effective churn prediction and identifying reasons why customers churn. These can be used to build effective customer retention strategies and maximize the profits of a business.

__1.2 Churn prediction using Artificial Intelligence__

Over the years, leveraging artificial intelligence(AI) has been able to show promising results in predicting customer churn by utilizing different techniques such as machine learning(ML) algorithms and data mining. These have provided significant benefits over traditional methods such as :

•	Being able to handle large amounts of data. This is important as large volumes are being generated from telecom industries and with different varieties.

•	Automating the process of analyzing and predicting customer churn, saving time, and reducing costs.

•	Analyzing complex and non-linear data on customer behavior can be crucial to finding predictors of customer churn such as customer location and usage patterns. These can allow companies to build better customer relations and target the right customers.

Some of the AI and machine learning(ML) techniques used in churn prediction involve using:

•	Supervised learning algorithms such as decision trees, random forest, and logistic regression help build predictive models. These models have been trained on past historical data with features such as customer age, usage patterns, and many more.

•	Unsupervised learning algorithms such as K-means clustering. These clustering algorithms are used to cluster data and identify customer churn patterns.

•	Deep learning (DL) techniques such as deep neural networks(DNN) for building predictive models with higher accuracy. These models can learn complex patterns with large amounts of customer data.

*Part B - Literature Review*

__2.1 Machine learning-based Churn prediction__
Traditional machine learning techniques such as logistic regression and decision trees have proven to be effective in terms of predicting customer churn. A study (Dalvi et al., 2016) used logistic regression(LR) and decision trees for churn prediction and to provide customer retention strategies.. The authors have also shown that using LR can help determine the important features and decision trees help in generating rules for customer retention strategies. LR is also computationally less expensive and requires less resources to setup.
However, these models may not be able to handle complex data and may not be able to accurately determine the important features for prediction. Also, LR can be very sensitive for any imbalance in the dataset which has not been mentioned in the research and be sensitive for outliers in the data.

__2.2 Deep learning-based Churn prediction__
The use of DL has shown tremendous progress in terms of classification problems. A study (Mishra and Reddy, 2017) have demonstrated that using Artificial Neural Network(ANN) can significantly improve accuracy compared to other existing ML models. The dataset consists of data from two different companies which contains historical data and other features. With the introduced architecture, the model was able to get an accuracy of 88%. The paper also highlights the use of the proposed model is also able to learn complex relationships which is crucial when dealing with large datasets that are being generated from telecom industries.
	But using neural networks(NN) can make it difficult to understand why a certain customer might churn making it less interpretable. Moreover, NN requires large amounts of data to train which may be a problem when it comes to smaller datasets, and they are highly complex and require high computational resources. They require a lot of hyperparameter tuning which can take time.


__2.3 Ensemble learning-based Churn prediction__
Several studies have proven using Random Forest(RF) for churn prediction can yield better accuracy compared to other classification algorithms. RF involves using an ensemble method that combines multiple decision trees to accurately make classifications.
A study by (Ullah et al., 2019) implemented and used RF to predict churn for a telecom company in South Asia. The authors have used RF here since it can handle large amounts of data and had a prediction accuracy of 85%. They were also able to provide retention strategies by identifying different features that were crucial to predict churn. They were able to use different feature engineering and feature selection techniques such as information gain and Attribute rank filtering.
Another study (Xie et al., 2009) highlights the need for a balanced dataset to train RF. The authors have addressed the data imbalance issue by using a random over-sampling technique and trained the model on this. This resulted in better performance and accuracy of the RF model with lower false positives.
However, these studies were based on one company, making it difficult to interpret how generalizable the models are. Also, the datasets being used have also not been revealed, making it difficult to evaluate the model’s scalability

In this project,  different machine learning models will be trained and be used to build a stacked ensemble classifier which is a type of ensemble learning method that combines the predictions of multiple base models to make a final prediction.
```python
import pandas as pd
import numpy as np
import matplotlib
from matplotlib import pyplot as plt
import matplotlib.ticker as mtick
```

```python
#import datsets
# dataset from kaggle https://www.kaggle.com/datasets/blastchar/telco-customer-churn
# remaned the csv as Telco_Customer_Churn for easy access
ds = pd.read_csv('Telco_Customer_Churn.csv')
ds.head()
```

```python
# conduct exploratory data analysis and feature engineering 
ds.dtypes
```

```python
# Getting rid of null values
ds.TotalCharges = pd.to_numeric(ds.TotalCharges, errors='coerce')
ds.dropna(inplace = True)
ds.isnull().sum()
```

```python
# Removing customerID from the dataframe
ds1 = ds.drop('customerID', axis='columns')
#Convert all categorical variables to numeric
ds1['gender'].replace({'Female' : 1, 'Male' : 0}, inplace=True)
ds1.replace('No internet service', 'No', inplace=True)
ds1.replace('No phone service', 'No', inplace=True)
col_convert = ['Partner', 'Dependents', 'PhoneService', 'OnlineSecurity', 'OnlineBackup', 'DeviceProtection', 'TechSupport',
              'StreamingTV', 'StreamingMovies', 'PaperlessBilling', 'Churn','MultipleLines']
for col in col_convert:
    ds1[col].replace({'Yes': 1, 'No': 0}, inplace=True)
    
# Get dummy data for some of the categorical data
df2 = pd.get_dummies(data = ds1, columns=['InternetService', 'Contract', 'PaymentMethod'])
df2.head()
```

```python
# All the features are now numerical data
df2.dtypes
```

```python
colors = ['#4D3425','#E4512B']
ax = (df2['Churn'].value_counts()*100.0 /len(df2)).plot(kind='bar',stacked = True,rot = 0,color = colors,figsize = (8,6))
ax.yaxis.set_major_formatter(mtick.PercentFormatter())
ax.set_ylabel('Customers',size = 10)
ax.set_xlabel('Churn',size = 10)
ax.set_title('Churn Visualization', size = 10)

totals = []
for i in ax.patches:
    totals.append(i.get_width())
total = sum(totals)

for i in ax.patches:
    ax.text(i.get_x()+.15, i.get_height()-4.0, \
            str(round((i.get_height()/total), 1))+'%',
            fontsize=12,
            color='white',
           weight = 'bold')
```

```python
#Scaling some of the data 
cols_scale = ['tenure', 'MonthlyCharges', 'TotalCharges']

from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()

df2[cols_scale] = scaler.fit_transform(df2[cols_scale])
df2.head()
```

```python
# Handling imbalanced data
X = df2.drop('Churn',axis='columns')
y = df2['Churn']

from imblearn.over_sampling import SMOTE
smote = SMOTE(sampling_strategy='minority')
X_sm, y_sm = smote.fit_resample(X,y)

#split data for training and testing
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X_sm,y_sm,test_size=0.3)

#find feature importance using information gain
# from sklearn.feature_selection import mutual_info_classif, SelectKBest
# selector = SelectKBest(score_func=mutual_info_classif, k=10)
# X_new = selector.fit_transform(X_sm, y_sm)
# X.head()
# info = mutual_info_classif(X_train, y_train)
# info = pd.Series(info)
# info.sort_values(ascending=False)
# info.sort_values(ascending=False).plot.bar(figsize=(20, 8))

#helper function to print classification report
from sklearn.metrics import classification_report, confusion_matrix, ConfusionMatrixDisplay,roc_auc_score, f1_score,accuracy_score,roc_curve

clf_name = []
roc_auc = []
f1 = []
def model_eval(clf, y_test, y_pred):
    print(classification_report(y_test, y_pred))
    cm=confusion_matrix(y_test, y_pred, labels=y_test.unique())
    disp = ConfusionMatrixDisplay(cm, display_labels=y_test.unique())
    disp.plot(cmap='cividis')
    m1 = roc_auc_score(y_test, y_pred)
    m2 = f1_score(y_test, y_pred)
    print('ROC_AUC_Score: {:.04f}'.format(m1))
    print('F1 Score: {:.04f}'.format(m2))
    clf_name.append(clf)
    roc_auc.append(m1)
    f1.append(m2)
```

__KNN Classifier__

```python
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier().fit(X_train, y_train)
y_pred = knn.predict(X_test)
clf = "KNN"
model_eval(clf,y_test, y_pred)
```

__SVM Classifier__

```python
from sklearn import svm
from sklearn.model_selection import GridSearchCV
# svm = svm.SVC().fit(X_train, y_train)
# model_eval(svm,y_test, y_pred)
svm2 = GridSearchCV(svm.SVC(gamma='auto'), {
    'C': [1,10,20],
    'kernel': ['rbf', 'linear']
},cv=3, return_train_score=False)

svm2.fit(X_train, y_train)
y_pred = svm2.predict(X_test)
clf = "SVM"
model_eval(svm2,y_test, y_pred)
```

__MLP Classifier__

```python
from sklearn.neural_network import MLPClassifier
mlp = MLPClassifier().fit(X_train, y_train)
mlp.score(X_test, y_test)
y_pred = mlp.predict(X_test)
clf = "MLP"
model_eval(clf,y_test, y_pred)
```

__Logistic Regression__

```python
from sklearn.linear_model import LogisticRegression
log = LogisticRegression()
log.fit(X_train, y_train)
y_pred = log.predict(X_test)
clf = "Logistic Regression"
model_eval(clf,y_test, y_pred)
```

__Random Forest__

```python
from sklearn.ensemble import RandomForestClassifier
rf = RandomForestClassifier()
#Define the parameter grid
param_grid = {
    'n_estimators': [10, 50, 100],
    'max_depth': [5, 10, 20],
    'min_samples_split': [2, 5, 10]
}
grid_search = GridSearchCV(estimator=rf, param_grid=param_grid, cv=5, n_jobs=-1)

# Fit the grid search to the data
grid_search.fit(X_train, y_train)

# Get the best parameters from the grid search
best_params = grid_search.best_params_

# Fit the optimized model to the data
rf_opt = RandomForestClassifier(**best_params).fit(X_train, y_train)

y_pred = rf_opt.predict(X_test)
clf = "Random Forest"
model_eval(clf,y_test, y_pred)
```

__XGBoost Classifier__

```python
import xgboost as xgb
gbm = xgb.XGBClassifier().fit(X_train,y_train)
y_pred = gbm.predict(X_test)
clf = "XGBoost"
model_eval(clf,y_test, y_pred)
```

__LGBM Classifier__

```python
 # %pip install lightgbm
from lightgbm import LGBMClassifier
lgbm_clf = LGBMClassifier().fit(X_train, y_train)
y_pred = lgbm_clf.predict(X_test)
clf = 'LightGBM'
model_eval(clf, y_test, y_pred)
```

__Stacked Classifier__

```python
from sklearn.ensemble import StackingClassifier

estimator_list = [
    ('knn', knn),
    ('rf',rf_opt),
    ('svm',svm2),
    ('mlp',mlp),
    ('xgb',gbm),
]

# Build stack model
stack_model = StackingClassifier(
    estimators=estimator_list, final_estimator=LogisticRegression()
)

stack_model.fit(X_train, y_train)
y_pred = stack_model.predict(X_test)
clf = "Stacked Model"
model_eval(clf,y_test, y_pred)
```

From the above implementation, the stacked ensemble classifier was able to outperform other algorithms. This can be further improved by fine tuning the hyper parameters 

*Part D - Testing*

One of the common techniques used for testing the performance of the model is to assess how well it will perform by splitting the dataset into training and test. This allows the model to be evaluated on data that it has not seen during training.

```python
print("Accuracy when training",stack_model.score(X_train, y_train))
print("Accuracy with test data", stack_model.score(X_test, y_test))
```

Another method to test the peformance of the model is to utilise the K fold cross validation which is a technique for evaluating the performance of a machine learning model by dividing the data into k subsets, or "folds", and training the model on k-1 of those folds while using the remaining fold as the validation set. This process is repeated k times, with each fold being used as the validation set once. The results are then averaged to give an estimate of the model's performance on unseen data.

Below is an implementation of K cross with 5 folds and the output shows the different accuracies of the model when training

```python
from sklearn.model_selection import cross_validate
def cross_validation(model, _X, _y, _cv=5):
      _scoring = ['accuracy', 'precision', 'recall', 'f1']
      results = cross_validate(estimator=model,
                               X=_X,
                               y=_y,
                               cv=_cv,
                               scoring=_scoring,
                               return_train_score=True)
      
      return {"Training Accuracy scores": results['train_accuracy'],
              "Mean Training Accuracy": results['train_accuracy'].mean()*100,
              "Training Precision scores": results['train_precision'],
              "Mean Training Precision": results['train_precision'].mean(),
              "Training Recall scores": results['train_recall'],
              "Mean Training Recall": results['train_recall'].mean(),
              "Training F1 scores": results['train_f1'],
              "Mean Training F1 Score": results['train_f1'].mean(),
              "Validation Accuracy scores": results['test_accuracy'],
              "Mean Validation Accuracy": results['test_accuracy'].mean()*100,
              "Validation Precision scores": results['test_precision'],
              "Mean Validation Precision": results['test_precision'].mean(),
              "Validation Recall scores": results['test_recall'],
              "Mean Validation Recall": results['test_recall'].mean(),
              "Validation F1 scores": results['test_f1'],
              "Mean Validation F1 Score": results['test_f1'].mean()
              }
cross_validation(stack_model, X_train, y_train)
```

*Part E - Evaluation*

The proposed work has been evaluated based on the confusion matrix.The confusion matrix table can assist in evaluating the performance of a classification model on a set of test data for which the true values are known. It is a powerful tool for analyzing the performance of a model, as it provides a detailed breakdown of the model's performance across all classes. The matrix is made up of four different values, true positives (TP), true negatives (TN), false positives (FP) and false negatives (FN). It can also allow to compute several evaluation metrics such as accuracy, precision, recall, and F1-score.
Accuracy: The overall accuracy of the model, which is the proportion of correct predictions (TP + TN) out of all predictions.
Precision: The proportion of true positives among all positive predictions (TP / (TP + FP))
Recall: The proportion of true positives among all actual positive cases (TP / (TP + FN))
F1-score: The harmonic mean of precision and recall (2 * (precision * recall) / (precision + recall))

```python
models = {
    'MLP': mlp,
    'LR': log,
    'SVM': svm2,
    'KNN': knn,
    'XGB': gbm,
    'LGBM':lgbm_clf,
    'RF': rf_opt,
    'Ensemble': stack_model
}

y_pred = {}
for name, model in models.items():
    y_pred[name] = model.predict(X_test)

results = {}
for name, y in y_pred.items():
    acc = accuracy_score(y_test, y)
    f1 = f1_score(y_test, y)
    results[name] = {'accuracy': acc, 'f1-score': f1}

# Plot the results
fig, (ax1, ax2) = plt.subplots(nrows=1, ncols=2, figsize=(10, 5))
ax1.set_title("Accuracy")
ax2.set_title("F1-score")

for name, scores in results.items():
    ax1.bar(name, scores['accuracy'])
    ax2.bar(name, scores['f1-score'])

plt.show()
```

These graphs helps to point out that ensemble methods can be a better choice for churn prediction as almost all of the simple machine learning algorithms such as logistic regression, support vector machines have been outperformed interms of accuracy and f1-score. The random forest was able to achieve the highest performance interms of accuracy reaching over 85% while the Light-gradient boosting machine goes around 83%. The stacked ensemble models have also achieved a higher accuracy and f1-score by combining all of the models to get an average score.

In conlusion, this project justifies  that using ensemble learning techniques for churn prediction can yield to higher performance than traditional methods. These techniques can also be improved using various hyper parameter optimization techniques and can be generalized for use with other datasets.
